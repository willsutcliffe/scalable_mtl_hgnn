```python
eff = []
rej = []
eff2 = []
rej2 = []
```

```python
bins = np.linspace(0,300,16)
```

```python
for i in range(0,len(bins)-1):
    dfcut = df.query(f"{bins[i]} < NumParticlesInEvent < {bins[i+1]}")
    dfcut2 = df2.query(f"{bins[i]} < NumParticlesInEvent < {bins[i+1]}")
    eff.append(np.mean(dfcut.NumSelectedParticlesFromHeavyHadronInEvent/dfcut.NumParticlesFromHeavyHadronInEvent))
    rej.append(np.mean(dfcut.NumSelectedBackgroundParticlesInEvent/dfcut.NumParticlesInEvent))
    eff2.append(np.mean(dfcut2.NumSelectedParticlesFromHeavyHadronInEvent/dfcut2.NumParticlesFromHeavyHadronInEvent))
    rej2.append(np.mean(dfcut2.NumSelectedBackgroundParticlesInEvent/dfcut2.NumParticlesInEvent))        
```

```python
eff
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define bin edges and values

# Create the step histogram
plt.step(bins[:-1], eff, where='mid', linewidth=2,label='Signal efficiency',color="black")
plt.step(bins[:-1], eff2, where='mid', linewidth=2,label='Signal efficiency',linestyle="dashed",color="black")
plt.step(bins[:-1], rej, where='mid', linewidth=2,label='Background retention',color="blue")
plt.step(bins[:-1], rej2, where='mid', linewidth=2,label='Background retention',color="blue",linestyle="dashed")
# Labels and formatting
plt.xlabel("Number of particles")
plt.ylabel("Efficiency")
#plt.title("Step Histogram")
plt.grid(True)
#plt.legend()
# Show the plot
plt.show()
```
