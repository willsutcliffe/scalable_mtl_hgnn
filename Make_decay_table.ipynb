```python
all_data = []
for decay in decays:
    data =[]
    #data.append(decay)
    for cat in categories:
        for model in models:
            signal_match = True
            if model != "dfei":
                print(f"../paper_results/{decay}_{model}/*signal_df.csv")
                signal_df = pd.read_csv(glob.glob(f"../paper_results/{decay}_{model}/*signal_df.csv")[0])
                if signal_match:
                    signal_df = signal_df.query("SigMatch == 1")
                else:
                    signal_df = signal_df
                sig_perfect_reco = 100 * len(signal_df.query('PerfectSignalReconstruction == 1')) / len(signal_df)
                sig_wrong_hierarchy = 100 * len(signal_df.query('AllParticles == 1')) / len(signal_df) - sig_perfect_reco
                sig_none_isolated = 100 * len(signal_df.query('NoneIso == 1')) / len(signal_df)
                sig_part_reco = 100 * (len(signal_df.query('PartReco == 1')) / len(signal_df) + len(
                    signal_df.query('NotFound == 1')) / len(signal_df))
                perf_numbers = pd.DataFrame(columns=["Decay","Model", "Perfect hierarchy", "Wrong hierarchy", "Not isolated", "Part reco"])
                perf_numbers = perf_numbers._append({"Decay": decay,"Model":model, "Perfect hierarchy": sig_perfect_reco,
                                                     "Wrong hierarchy": sig_wrong_hierarchy, "Not isolated": sig_none_isolated,
                                                     "Part reco": sig_part_reco}, ignore_index=True)
                data.append(perf_numbers[cat].values[0])
            else:
                print(f"/home/sutclw/Work/Zurich/LHCb/GNNs/lhcbdfei/inference/python/dfei_paper_results/{decay}_sig_df.csv")
                signal_df = pd.read_csv(f"/home/sutclw/Work/Zurich/LHCb/GNNs/lhcbdfei/inference/python/dfei_paper_results/{decay}_sig_df.csv")
                if signal_match and decay != "inclusive":
                    signal_df = signal_df.query("SignalMatch == 1")
                else:
                    signal_df = signal_df
                sig_perfect_reco = 100 * len(signal_df.query('PerfectSignalReconstruction == 1')) / len(signal_df)
                sig_wrong_hierarchy = 100 * len(signal_df.query('QuasiSignalReconstruction == 1')) / len(signal_df) 
                sig_none_isolated = 100 * len(signal_df.query('PartialSignalReconstruction == 1')) / len(signal_df)
                sig_part_reco = 100  - sig_perfect_reco - sig_wrong_hierarchy - sig_none_isolated
                perf_numbers = pd.DataFrame(columns=["Decay","Model", "Perfect hierarchy", "Wrong hierarchy", "Not isolated", "Part reco"])
                perf_numbers = perf_numbers._append({"Decay": decay,"Model":model, "Perfect hierarchy": sig_perfect_reco,
                                                     "Wrong hierarchy": sig_wrong_hierarchy, "Not isolated": sig_none_isolated,
                                                     "Part reco": sig_part_reco}, ignore_index=True)
                data.append(perf_numbers[cat].values[0])

    all_data.append(data)

    
```
